 ### A note on consumer price index(CPI): When calculating the consumer price index, the final consumer price index result represents the average change in prices that consumers will spend on a basket of goods and services over time. This is how economic and financial professionals identify and determine inflation. They can then use CPI to determine the economy's aggregate price levels to measure the purchasing price of an entire country or a specific area. The consumer price index formula is:

### Cost of products or services in a current period/cost of products or services in a previous time period x 100 = consumer price index

### A note on price elasticity of demand for necessary goods: According to economic theory, the price elasticity of demand for necessary goods is less than unity(Demand is price inelastic for necessary goods).

```python
df_200= df_2[(df_2['Store']!=17)&(df_2['Store']!=33)&(df_2['Store']!=36)&(df_2['Store']!=37)&(df_2['Store']!=38)&(df_2['Store']!=42)&(df_2['Store']!=43)&(df_2['Store']!=44)]
```

```python
## time-series forecasting on df_200:
```

```python
plt.figure(figsize=(10,7))
plt.plot(df_200['Date'],df_200['Weekly_Sales'])
plt.grid()
plt.show()
```

```python
# train-test split:
df_200_tr= df_200[df_200['Date']< '2011-05-1']
```

```python
df_200_tr # train data-set
```

```python
df_200_tst= df_200[df_200['Date']>= '2011-05-1']
```

```python
df_200_tst # test data-set
```

```python
from pmdarima import auto_arima
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
a_arima= auto_arima(df_200_tr['Weekly_Sales'],trace= True,seasonal= True)
```

```python
arima_tr= ARIMA(df_200_tr['Weekly_Sales'],order=(5,0,5)) 
model= arima_tr.fit()
```

```python
model.summary() # model trained
```

```python
# testing how good my model is forecasting:
test_forecast=model.predict(start= len(df_200_tr),end=len(df_200_tst)+len(df_200_tr)-1,dynamic=False)
```

```python
test_forecast # forecasted values
```

```python
df_200_tst # real values
```

```python
df_200_tst
```

```python
test_forecast.index
```

```python
df_200_tst.set_index(test_forecast.index,inplace=True)
```

```python
df_200_tst
```

```python
df_200_tst['Weekly_sales_prediction']= test_forecast
```

```python
df_200_tst
```

```python
plt.figure(figsize=(10,8))
plt.plot(df_200_tst['Date'],df_200_tst['Weekly_Sales'])
plt.plot(df_200_tst['Date'],df_200_tst['Weekly_sales_prediction'])
plt.grid()
plt.show()
```
